### Reflexion Agent
A Reflexion Agent is a type of intelligent agent that learns from its own reflections or past experiences. Unlike a simple reflex agent that reacts instantly to the current percept, a reflexion agent:

Generates an action or response.

Reflects on that action — evaluates success, mistakes, or outcomes.

Stores reflections (like a memory or log).

Uses those reflections in future runs to improve performance.

In short: Reflexion agents “think about what they did” and use it to act better next time.
